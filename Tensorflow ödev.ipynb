```python
#soru1
data=pd.read_csv("adult.csv")
df=data.copy()

df_workclass=df["workclass"].unique()
df_workclass=df_workclass[df_workclass!='?']


df_occupation=df["occupation"].unique()
df_occupation=df_occupation[df_occupation!='?']


df_nativecountry=df["native-country"].unique()
df_nativecountry=df_nativecountry[df_nativecountry!='?']


for i in range(df.shape[0]):
    n_workclass=int(random.random()*df_workclass.shape[0])
    n_occupation=int(random.random()*df_occupation.shape[0])
    n_nativecountry=int(random.random()*df_nativecountry.shape[0])
    if df.loc[i,"workclass"]=='?':
        df.loc[i,"workclass"]=df_workclass[n_workclass]
    if df.loc[i,"occupation"]=='?':
        df.loc[i,"occupation"]=df_occupation[n_occupation]
    if df.loc[i,"native-country"]=='?':
        df.loc[i,"native-country"]=df_nativecountry[n_nativecountry]    

print(df)        

```
